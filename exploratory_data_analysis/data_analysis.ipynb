```python
import pandas as pd
import csv

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
from itertools import chain
import math

# use geopy API to get coordinates from address
from geopy.geocoders import Nominatim
```

```python
# import csv file of 2022 into dataframe
dataset2022 = pd.read_csv('../datasets/medals2022.csv', usecols=['medal_type', 'athlete_name', 'event', 'discipline', 'country'])

# add year - 2022
dataset2022.insert(dataset2022.shape[1], 'slug_game', '2022-beijing')

# convert the medal type to all upper style
dataset2022['medal_type'] = dataset2022['medal_type'].str.upper()

# adjust the order of coloumns
order2022 = ['slug_game', 'discipline', 'event', 'medal_type', 'athlete_name', 'country']
dataset2022 = dataset2022[order2022]

dataset2022.info()
```

```python
# adjust the last name and the first name to have a unified format
full_name = dataset2022['athlete_name']
full_name_split = full_name.str.split(' ',expand=True)

array = [0,0,0,0]
for index, row in full_name_split.iterrows():
    for i in range(0, 4):
        if row[i] == None:
            array[i] = 1
        elif row[i] == row[i].title():
            array[i] = 2
        else:
            array[i] = 3
    
    if array == [3, 2, 1, 1]:
        full_name.loc[index] = row[1] + ' ' + row[0]
    if array == [3, 2, 2, 1]:
        full_name.loc[index] = row[1] + ' ' + row[2] + ' ' + row[0]
    if array == [3, 3, 2, 1]:
        full_name.loc[index] = row[2] + ' ' + row[0] + ' ' + row[1]    
    if array == [3, 3, 3, 2]:
        full_name.loc[index] = row[3] + ' ' + row[0] + ' ' + row[1] + ' ' + row[2]
        
    array = [0,0,0,0]
        
dataset2022['athlete_name'] = full_name
```

```python
# import csv file of 1896 into dataframe
dataset1896 = pd.read_csv('../datasets/medals1986_2020.csv', usecols=['discipline_title', 'slug_game',  'event_title', 'medal_type', 'athlete_full_name', 'country_name'])

# change the place of the city name and the year, for easy sort
#dataset1896['slug_game'] = dataset1896['slug_game'].str.split('-',expand=True)[1] + '-' + dataset1896['slug_game'].str.split('-',expand=True)[0]
# replace undefined symbol by '
dataset1896['event_title'] = dataset1896['event_title'].str.replace('â€™', '\'')

# adjust the order of coloumns
order1896 = ['slug_game', 'discipline_title', 'event_title', 'medal_type', 'athlete_full_name', 'country_name']
dataset1896 = dataset1896[order1896]

# drop rows without names
dataset1896 = dataset1896.dropna(axis=0,how='any').reset_index(drop=True) 

# change the coloumns' titles
dataset1896.columns = ['slug_game', 'discipline', 'event', 'medal_type', 'athlete_name', 'country']

dataset1896.info()
```

```python
# change the name format
slug_game = dataset1896['slug_game']
slug_game_split = dataset1896['slug_game'].str.split('-',expand=True)

host_year_name = ''
for index, row in slug_game_split.iterrows():
    for i in range(3,-1,-1):
        if row[i] != None:
            host_year_name = row[i]
            for j in range(0, i):
                host_year_name = host_year_name + '-' + row[j]
            break
    slug_game.loc[index] = host_year_name
    host_year_name = ''

dataset1896['slug_game'] = slug_game
```

```python
# combine the two dataset together
dataset = pd.concat([dataset2022, dataset1896]).reset_index(drop=True)

# remove all summer olympics
dataset = dataset.drop(dataset[(dataset['slug_game'] == '1896-athens') | (dataset['slug_game'] == '1900-paris') | (dataset['slug_game'] == '1904-st-louis') | (dataset['slug_game'] == '1908-london') | (dataset['slug_game'] == '1912-stockholm') | (dataset['slug_game'] == '1920-antwerp') | (dataset['slug_game'] == '1924-paris') | (dataset['slug_game'] == '1928-amsterdam') | (dataset['slug_game'] == '1932-los-angeles') | (dataset['slug_game'] == '1936-berlin') | (dataset['slug_game'] == '1948-london') | (dataset['slug_game'] == '1952-helsinki') | (dataset['slug_game'] == '1956-melbourne') | (dataset['slug_game'] == '1960-rome') | (dataset['slug_game'] == '1964-tokyo') | (dataset['slug_game'] == '1968-mexico-city') | (dataset['slug_game'] == '1972-munich') | (dataset['slug_game'] == '1976-montreal') | (dataset['slug_game'] == '1980-moscow') | (dataset['slug_game'] == '1984-los-angeles') | (dataset['slug_game'] == '1988-seoul') | (dataset['slug_game'] == '1992-barcelona') | (dataset['slug_game'] == '1996-atlanta') | (dataset['slug_game'] == '2000-sydney') | (dataset['slug_game'] == '2004-athens') | (dataset['slug_game'] == '2008-beijing') | (dataset['slug_game'] == '2012-london') | (dataset['slug_game'] == '2016-rio') | (dataset['slug_game'] == '2020-tokyo')].index).reset_index(drop=True)
```

```python
# import country-capital dictionary file (csv)
country_cap = pd.read_csv('../datasets/country_capital.csv', usecols=['country', 'capital'])

# convert df to dictionary
dict_cap = dict(zip(country_cap['country'],country_cap['capital']))

# map the country name to its capital, if not matched (NaN), then still use the country name
df_count_by_country['address'] = df_count_by_country['country'].map(dict_cap)
df_count_by_country['address'].fillna(df_count_by_country['country'],inplace=True)

# add all rows that have the same address together
df_count_by_address = df_count_by_country.drop(columns=['country']).groupby('address').sum()

# use the number of medals of each country to calculate the point size in the world map
log_num = df_count_by_address['numbers'].map(lambda x: 5 * math.log(x+1,10))
```

```python
geolocator=Nominatim(user_agent="data_preprocessing")
array_loc = []
for index, row in df_count_by_address.iterrows():
    array_loc.append(geolocator.geocode(index))
    
```

```python
# draw the world map
def draw_map(m, scale=0.2):
    m.shadedrelief(scale=scale)
    
    lats = m.drawparallels(np.linspace(-90, 90, 13))
    lons = m.drawmeridians(np.linspace(-180, 180, 13))
    
    lat_lines = chain(*(tup[1][0] for tup in lats.items()))
    lon_lines = chain(*(tup[1][0] for tup in lons.items()))
    all_lines = chain(lat_lines, lon_lines)
    
    for line in all_lines:
        line.set(linestyle='-', alpha=0.3, color='w')

# add points based on coordinates and point size (based on # medals)
def draw_point(m, x, y, num, name):
    x, y = m(x, y)
    plt.plot(x, y, 'ok', markersize=num, color='blue')
        
        
fig = plt.figure(figsize=(20, 15), edgecolor='w')
m = Basemap(projection='cyl', resolution=None, llcrnrlat=-90, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180,)
draw_map(m)

for i in range(0,40):
    if array_loc[i] != None: # some row don't have coordinates, e.g. Olympic Athletes
        draw_point(m, array_loc[i].longitude, array_loc[i].latitude, log_num[i], array_loc[i].address)

plt.show()
```
